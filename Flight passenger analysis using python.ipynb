### Import all the datasets.

```python
pass_data=pd.read_csv('passenger data.csv')
surv_data=pd.read_csv('survey data.csv')
```

### Create a new dataframe “df” by joining the 2 datasets


```python
df=pd.merge(pass_data, surv_data, how='left', left_on='id', right_on='id')
```

### Set “ID” as Index columns

```python
df=df.set_index('id')
```

### Deal for missing values

```python
# Drop the null values
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

###  Drop the duplicate values in the df.

```python
df=df.drop_duplicates()
df.head()
```

### Rename the columns with string title format 

```python
df.columns
```

```python
df.columns = df.columns.str.title()
```

```python
df.columns
```

### Replace the Column values in “Satisfaction” column with Logic

```python
df=df.replace(['satisfied'],'Yes')
df=df.replace(['neutral or dissatisfied'],'No')
df.head()
```

## Create a new Dataframe “newdata” which contains all numerical columns as well as dummy values of the categorical columns. Perform Correlation on this dataframe
